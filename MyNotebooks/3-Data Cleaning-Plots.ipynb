## Histogramas de frecuencias

La regla de Sturges indica el número de clases(divisiones) que se deben considerar al elaborarse un histograma:

    c=1+log2(M), siendo M el tamaño de la muestra

```python
sturges=int(np.ceil(np.log2(3333)))
plt.hist(data["Day Calls"],bins=sturges)#bins indica el número de divisiones de los trozos
plt.xlabel("Número de llamadas al día")
plt.ylabel("Frecuencia")
plt.title("Histograma del número de llamadas al día")
```

```python
plt.hist(data["Day Calls"],bins=[0,60,90,120,150,180])
plt.xlabel("Número de llamadas al día")
plt.ylabel("Frecuencia")
plt.title("Histograma del número de llamadas al día")
```

## Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("Número de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```

   * Por debajo del borde inferior de la caja quedan un cuarto de los datos(percentil 25)
   * Por encima del borde superior de la caja quedan el 25% de los datos(percentil 75)
   * La raya de color amarillo es la mediana o el percentil 50
   * La altura de la caja es el rango intercuartilo(IQR)
   * Las barras delimitadoras de los circulitos(bigotes) estan a 1.5*IQR

```python
IQR=data["Day Calls"].quantile(0.75)-data["Day Calls"].quantile(0.25)
```

```python
#Bigote inferior
data["Day Calls"].quantile(0.25)-1.5*IQR
```

```python
#Bigote superior
data["Day Calls"].quantile(0.75)+1.5*IQR
```
