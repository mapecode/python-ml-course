# Variables Dummy

```python
data=titanic
data["sex"].head(3)#Variable categorica de la que se obtendran dos variables dummy(male y female)
```

```python
dummy_sex=pd.get_dummies(data["sex"],prefix="sex")
dummy_sex.head(3)
```

```python
#Eliminar la columna categorica sex y crear dos nuevas columnas generadas de las variables dummy:
data2=data.drop(["sex"],axis=1)#Eliminar columna
pd.concat([data2,dummy_sex],axis=1).head(2)#Concatenar las nuevas columnas
```

```python
def createDummies(dataFrame,var_name,axis):
    #pd => pandas library
    dummy=pd.get_dummies(dataFrame[var_name],prefix=var_name)
    dataFrame=dataFrame.drop(var_name,axis)
    dataFrame=pd.concat([dataFrame,dummy],axis)
    return dataFrame
```

```python
createDummies(titanic,"sex",1).head(3)
```
